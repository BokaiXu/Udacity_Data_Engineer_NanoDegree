```python
# Airport dataset
df_airport=pd.read_csv('airport-codes_csv.csv')

# Function to split the coordinates
def split_coord1(Series):
    return float(Series.split()[0][:-1])

def split_coord2(Series):
    return float(Series.split()[1])

df_airport['coordinate1']=df_airport['coordinates'].apply(lambda x: split_coord1(x))
df_airport['coordinate2']=df_airport['coordinates'].apply(lambda x: split_coord2(x))

# Select United States airport
df_airport=df_airport[df_airport['iso_country']=='US']

# Drop unclear columns
df_airport.drop(['ident','elevation_ft','continent','iso_country','gps_code','iata_code','coordinates'],axis=1,inplace=True)

df_airport.rename(columns={'local_code':'i94port'},inplace=True)
df_airport.sample(5)
```

#### 4.3 Data dictionary 
##### Fact_table:    
cicid-primary key, i94mon-the month the immigrant enter United States, visa type-the type of visa the immigrant has, gender-the gender of the immigrant, dtadfile-the date this i94 was generated, i94res-foreign key, i94addr-foreign key, i94port-foreign key    
##### dim_city:    
i94addr-primary key(which state the immigrant enter), male population-number of male in the state, female population-number of female in the state, total population-number of total people in the state, number of veterans-number of veterans in the state, foreign born-number of foreign born in the state, count-number of count    
##### dim_temperature:    
i94res-primary key(which country the immigrant come from), country-country of the immigrant, city-city of the country, average temperature-the average temperature of a month, uncertainty-the uncertainy of the average temperature, month-from 1 to 12    
##### dim_airport:    
i94port-primary key(which airport the immigrant land on), type-type of airport, name-name of the airport, iso_region-which state of US, municipality-which city control this airport, coordinate1-the coordinate of the airport, coordinate2-the coordinate of the airport
